```python
# a - Create a Dataframe to store Student Name and Their 4 Subject Marks 

import pandas as pd 
```

```python

data = {'Cloud':[50 , 60 , 40 , 50 ] , 
        'Crypto':[40 , 50 , 40 , 50] , 
        'WSN' : [ 40 , 45 , 50 , 55] , 
        'DV' : [ 35 , 30 , 40 , 60]}

data
```

```python
index = ['Tushar' , 'Shreyas' , 'Akshay' , 'Junaid']

df = pd.DataFrame(data , index = index)
df
```

```python
#b - add a new column Total To maintan Total Marks as per Student 

total = df['Cloud'] + df['Crypto'] + df['WSN'] + df['DV']

df['Total'] = total
df
```

```python
#c - Add a new column average to maintain avg marks per subject 

avg = df['Cloud'] + df['Crypto'] + df['WSN'] + df['DV']/4

df['Average'] = avg

df
```
