```python
#a - create a dataFrame called df from the following ['a', 'b', 'c', 'd', 'e', 'f', 'g']


import pandas as pd
import numpy as np
```

```python
index = ['a', 'b', 'c', 'd', 'e', 'f', 'g' , 'h']

data = {
    'Animal' :[ 'cat' , 'cat' , 'dog' , 'cat' , 'dog' , 'rabbit' , 'rabbit' , 'dog'],
    'Age' : [2.5 , 3.5 , 1.2 ,np.nan , 1.5 , np.nan ,  0.5 , 2.0],
    'Visit' : [1 , 3 , 2 , 3 , 2 , 3 , 1 , 1],
    'priority' :['yes' , 'yes' , 'no' , 'yes' , 'no' , 'no' , 'no' , 'yes']
}

df = pd.DataFrame(data , index = index)
df
```
